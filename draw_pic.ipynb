```python
import matplotlib.pyplot as plt
import numpy as np

epochs = list(range(1, 21))
train_mse = [0.9539, 0.8350, 0.8178, 0.8032, 0.7889,
             0.7718, 0.7554, 0.7391, 0.7264, 0.7157,
             0.7074, 0.7007, 0.6942, 0.6890, 0.6850,
             0.6804, 0.6767, 0.6733, 0.6699, 0.6670]
valid_mse = [0.8469, 0.8301, 0.8194, 0.8099, 0.8004,
             0.7876, 0.7798, 0.7687, 0.7689, 0.7727,
             0.7616, 0.7626, 0.7619, 0.7633, 0.7638,
             0.7615, 0.7658, 0.7673, 0.7679, 0.7699]

color1 = 'b'  # 蓝色
color2 = 'r'  # 红色

# 设置好看的配色
color1 = 'b'  # 蓝色（train MSE）
color2 = 'r'  # 红色（valid MSE）

# 绘制训练和验证MSE折线图
plt.figure(figsize=(8, 6))
plt.plot(epochs, train_mse, label='Train MSE', color=color1)
plt.plot(epochs, valid_mse, label='Valid MSE', color=color2)

# 添加图例
plt.legend()

# 添加标题和标签
plt.title('Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('MSE')

# 显示图形
plt.grid(True)
plt.show()
```
